```python
#앤스콤 데이터 집합 불러오기
anscombe = sns.load_dataset("anscombe")
print(anscombe)
print(type(anscombe))
```

```python
#matplotlib라이브러리로 그래프 그리기
%matplotlib notebook
import matplotlib.pyplot as plt

dataset_1 = anscombe[anscombe['dataset']=='I']
```

```python
# plot 매서드 - 선그래프
plt.plot(dataset_1['x'],dataset_1['y'])

# plot 매서드 - 점그래프
plt.plot(dataset_1['x'],dataset_1['y'],'o')
```

### [히스토그램]

```python
tips = sns.load_dataset("tips")
print(tips.head())
```

```python
fig = plt.figure()
axes1 = fig.add_subplot(1,1,1)

axes1.hist(tips['total_bill'],bins=10) #x축 간격 bins
axes1.set_title('Histogram of Total Bill') #그래프 제목
axes1.set_xlabel("Frequency") #x축 제목
axes1.set_ylabel("Total Bill") #y축 제목
```

### [산점도]

```python
scatter_plot = plt.figure()

axes1 = scatter_plot.add_subplot(1,1,1)
axes1.scatter(tips['total_bill'],tips['tip'])

axes1.set_title('Scatterplot of Total Bill vs Tip')
axes1.set_xlabel('Total Bill')
axes1.set_ylabel('Tip')
```

### [박스 그래프]

```python
boxplot = plt.figure()

axes1 = boxplot.add_subplot(1,1,1)

axes1.boxplot([tips[tips['sex'] == 'Female']['tip'],
              tips[tips['sex'] == 'Male']['tip']],
              labels=['Female','Male'])

axes1.set_xlabel('Sex')
axes1.set_ylabel('Tip')
axes1.set_title('Boxplot of Tips by Sex')
```

### [산점도 - 다변량 데이터]

```python
def recode_sex(sex):
    if sex == 'Female':
        return 0
    else:
        return 1
```

```python
tips['sex_color'] = tips['sex'].apply(recode_sex)

scatter_plot = plt.figure()

axes1 = scatter_plot.add_subplot(1,1,1)
axes1.scatter(
    x = tips['total_bill'],
    y = tips['tip'],
    s = tips['size']*10,
    c = tips['sex_color'],
    alpha = 0.5)

axes1.set_title('Total Bill vs Tip Colored by Sex and Sized by Size')
axes1.set_xlabel('Total Bill')
axes1.set_ylabel('Tip')
```

## seabon 라이브러리

```python
import seaborn as sns

tips = sns.load_dataset("tips")
```

```python
ax = plt.subplots()
ax = sns.distplot(tips['total_bill'])
ax.set_title('Total Bill Histogram with Desnsity Plot')
```

### seabon 라이브러리로 다변량 그래프 그리기

```python
ax = plt.subplots()

#바이올린 그래프
ax = sns.violinplot(x='time',y='total_bill',hue='sex',data=tips,split=True)
```

```python
# 산점도 그래프
scatter = sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', fit_reg=False)
```

```python
# 관계 그래프
fig = sns.pairplot(tips, hue='sex')
```
